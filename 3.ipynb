#**seccion 1: Limpieza y transformacion de datos**

```python
#no hay valores nulos
df.isnull().sum()
```

```python
#analisis por cuartiles
q1 = df["Cantidad_Migrantes"].quantile(0.25)
q3 = df["Cantidad_Migrantes"].quantile(0.75)
iqr = q3 - q1

lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

df_filtrado = df[(df["Cantidad_Migrantes"] >= lim_inf) & (df["Cantidad_Migrantes"] <= lim_sup)]
df_filtrado.head()
#no tiene outliers segun la regla intercuartilica
```

```python
df["Razon_Migracion"].unique()
```

```python
#Reemplazo de valores en "Razon_mmigracion" con un mapeo
mapeo = {
    "Económica" : "Trabajo",
    "Conflicto" : "Guerra",
    "Educativa" : "Reunificacion"
}

df_filtrado = df_filtrado.replace({"Razon_Migracion": mapeo})
df_filtrado.head()
```

```python
#valores perdidos
df_filtrado.isnull().sum()
#no hay nulos, sin valores perdidos
```
